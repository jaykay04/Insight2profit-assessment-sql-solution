```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from fake_useragent import UserAgent
```

```python
url = "https://www.espn.com/college-football/stats/player"
ua = UserAgent()
userAgent = ua.random
headers = {"user-agent": userAgent}
page = requests.get(url, headers = headers)


soup = BeautifulSoup(page.content, "html.parser")
```

```python
espn_table = soup.find("div", class_ = "ResponsiveTable ResponsiveTable--fixed-left mt4 Table2__title--remove-capitalization")
headers = [header.text.strip() for header in espn_table.find_all("th")]

name_rows = []
stat_rows = []

# Step 1: Get the first tbody (player names)
name_tbody = espn_table.find_all("tbody")[0]
for row in name_tbody.find_all("tr"):
    fixed = [th.text.strip() for th in row.find_all("td")]
    name_rows.append(fixed)

# Step 2: Get the second tbody (stats)
stat_tbody = espn_table.find_all("tbody")[1]
for row in stat_tbody.find_all("tr"):
    stats = [td.text.strip() for td in row.find_all("td")]
    stat_rows.append(stats)

# Step 3: Combine them row-wise
data_rows = []
for name, stats in zip(name_rows, stat_rows):
    data_rows.append(name + stats)
    
    
espn_df = pd.DataFrame(data_rows, columns = headers)
espn_df
```
