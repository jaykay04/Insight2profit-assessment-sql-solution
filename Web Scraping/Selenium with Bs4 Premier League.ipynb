```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import pandas as pd

options = Options()
#options.add_argument("--headless")  #This runs in headless mode
options.add_argument("--disable-gpu")
#options.add_argument("window-size = 1200, 800")
#options.add_argument("--start-maximized")
#options.add_argument("--incognito")


service = Service(executable_path = r'C:\Users\LENOVO\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe')
driver = webdriver.Chrome(service = service, options = options)


driver.get("https://www.premierleague.com/players")
time.sleep(5)


while True:
    last_height = driver.execute_script("return document.body.scrollHeight")
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    if new_height == last_height:
        break
time.sleep(10)
        

page = driver.page_source
soup = BeautifulSoup(page, 'html.parser')

table = soup.find("table")

#Scrape data into Dataframe
if table:
    headers = [th.text.strip() for th in table.find_all("th")]
    players_table = pd.DataFrame(columns = headers)
    
    #Extract Rows
    for j in table.find_all("tr")[1:]: # This skips the first header which is actually the first row
        row_data = j.find_all("td")
        row = [td.text.strip() for td in row_data]
        
        #Append row only if it matches header length
        if len(row) == len(headers):
            players_table.loc[len(players_table)] = row
            
    print(players_table)
else:
    print("Table Not Found!!!")
```
